Сделайте кластеризацию данных методом `KMeans` (собственная имплементация и готовая реализация), обучив модель на данных со сниженной за счет PCA размерностью. В данном случае мы подскажем, что нужно искать именно 6 кластеров, но в общем случае мы не будем знать, сколько кластеров надо искать.

Параметры:

- **n_clusters** = n_classes (число уникальных меток целевого класса)
- **n_init** = 100
- **random_state** = RANDOM_STATE (для воспроизводимости результата)

Остальные параметры со значениями по умолчанию.

````python
import sys


class MyKMeans:
    def __init__(self, n_clusters, n_init=100, random_state=RANDOM_STATE, max_iter=100):
        self.n_clusters = n_clusters
        self.n_init = n_init
        np.random.seed(random_state)
        self.max_iter=max_iter

        self.inert = sys.float_info.max
        self.lables = None
        self.centers = None

    def fit(self, X):
        self.inert = sys.float_info.max
        self.lables = None
        self.centers = None
        for _ in range(self.n_init):
            centres = X[np.random.choice(X.shape[0], self.n_clusters)]
            for _ in range(self.max_iter):
                dist = np.sqrt(((X - centres[:, np.newaxis]) ** 2).sum(axis=2))
                labels = np.argmin(dist, axis=0)
                
                new_centres = np.array([X[labels==i].mean(axis=0) for i in range(self.n_clusters)])
                if (np.abs(centres - new_centres) < 1e-6).all():
                    break
                centres = new_centres

            inert = sum(np.min(((X - centres[:, np.newaxis]) ** 2).sum(axis=2), axis=0))
            if inert < self.inert:
                self.inert = inert
                self.lables = labels
                self.centers = centres

    def predict(self, X):
        dist = np.sqrt(((X - self.centers[:, np.newaxis])**2).sum(axis=2))
        return np.argmin(dist, axis=0) 

    def fit_predict(self, X):
        self.fit(X)
        return self.predict(X)
````

````python
model = MyKMeans(n_clusters=n_classes, n_init=100, random_state=RANDOM_STATE)
cluster_labels = model.fit_predict(X_pca)
````
